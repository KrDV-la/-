## Изменение названий столбцов

Приведем название колонок к змеиному стилю

```python
data.columns
```

```python
import re
```

```python
def camel_to_snake(name):
    return re.sub(r'(?<!^)(?=[A-Z])', '_', name).lower()
```

```python
new_columns = {}
for i in data.columns:
    new_columns[i] = camel_to_snake(i)
data.rename(columns=new_columns, inplace=True)
```

```python
data.columns
```

## Удаление столбцов

Сократим столбцы и удалим ненужные для нашего исследования или почти пустые

```python
dlt_columns = ['gross_tertiary_education_enrollment', 'gross_primary_education_enrollment_country', 
               'cpi_country', 'cpi_change_country', 'life_expectancy_country', 'status', 'date', 'birth_date', 
               'organization', 'birth_day', 'longitude_country', 'latitude_country', 'industries',
               'tax_revenue_country_country', 'state', 'residence_state_region', 'title', 'rank', 'source']
```

```python
data.drop(labels = dlt_columns, axis = 1, inplace = True)
data
```

## Изучение пропущенных значений

Проверим количество пустых значений в столбцах

```python
data.isnull().sum()
```

Проверим количество пустых значений в столбцах в **процентах**

```python
(data.isnull().sum() / len(data)) * 100
```

Заменим пустые значения

```python
# заменяем средним значением, так не сбиваются средние значения
data['age'] = data['age'].fillna(data['age'].mean()) 
data['birth_year'] = data['birth_year'].fillna(data['birth_year'].mean())
data['total_tax_rate_country'] = data['total_tax_rate_country'].fillna(data['total_tax_rate_country'].mean())
data['population_country'] = data['population_country'].fillna(data['population_country'].mean())
```

```python
# заполняем пустотой, чтобы программа работала исправно
data['country'] = data['country'].fillna('')
data['city'] = data['city'].fillna('')
data['first_name'] = data['first_name'].fillna('')
```

```python
# вставляем ноль, чтобы он не влиял на статистику(месяцы с 1 до 12)
data['birth_month'] = data['birth_month'].fillna(0)
```

```python
# делаем в стиле значений в столбце, ноль ставим для обозначения пустой колонки
data['gdp_country'] = data['gdp_country'].fillna('$ 0')
```

```python
data.isnull().sum()
```

**Все пропуски заполнены**

## Изменение типов данных

```python
data.dtypes
```

Изменим тип данных у столбцов **age, birth_year, birth_month, population_country** с float64 на int64 и столбца **gdp_country** c object на int64

```python
data = data.astype({'age': 'int64', 'birth_year': 'int64', 'birth_month': 'int64', 'population_country': 'int64'})
```

```python
# убираем $ и запятые из цифр
cnt = 0
for i in data.gdp_country:
    data.at[cnt, 'gdp_country'] = i.strip().replace(',', '')[1:] 
    cnt += 1
```

```python
data.gdp_country = data.gdp_country.astype('int64')
```

## 2. Диаграммы, библиотека Seaborn

```python
import seaborn as sns  
```

```python
# Подсчет количества миллиардеров по возрасту  
age_counts = data['age'].value_counts().reset_index()  
age_counts.columns = ['age', 'count']  

# Построение диаграммы рассеяния  
sns.scatterplot(data=age_counts, x='age', y='count', size=0.5)  
plt.title('Количество миллиардеров по возрасту')  
plt.xlabel('Возраст')  
plt.ylabel('Количество миллиардеров')  
plt.show()
```

```python
# Фильтрация данных в пределах возраста от 55 до 68 лет
df_filtered = data[(data['age'] >= 55) & (data['age'] <= 68)]

# Группировка данных по возрасту, полу и self-made статусу
df_grouped = df_filtered.groupby(['age', 'gender', 'self_made']).size().reset_index(name='count')

# Построение графика
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Построение столбчатой диаграммы с разделением по полу и self-made статусу
sns.barplot(data=df_grouped, x='age', y='count', hue='self_made', 
            dodge=True, palette='Set1', errorbar=None)

# Настройка меток и легенды
plt.title("Количество миллиардеров по возрастам от 55 до 68 лет с учетом self-made статуса")
plt.ylabel("Количество миллиардеров")
plt.xlabel("Возраст")
plt.legend(title='Self-made статус', loc='upper right')

# Показ графика
plt.show()

```
